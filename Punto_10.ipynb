```python
"""Empezamos abriendo los archicos txt en 3 variables distintas, utilizamos el "r" para establecer que solo vamos a 
leer el texto y lo decodificamos a utf-8 para que pueda ser leido correctamente
"""
txtnombre = open("nombres_1.txt", "r", encoding="utf-8")
txtnotas1 = open("eval1.txt", "r", encoding="utf-8")
txtnotas2 = open("eval2.txt", "r", encoding="utf-8")

"""Mediante el .read() leemos el contenido de cada uno de las variables y con el split lo hacemos una lista
"""
nombres= txtnombre.read().split(' ')
notas1= txtnotas1.read().split(',')
notas2= txtnotas2.read().split(',')

"""Luego de inicializar una lista en blanco en la cual vamos a introducir las otras 3 hechas previamente
tambien inicializamos una variable promedio para ir acumulando las notas"""

estructura=[]

promedio=0

"""Mediante el .zip() juntamos las 3 listas y recorremos de manera rapida con otras 3 variables las cuales vamos a guardar
en un diccionario auxiliar para luego a√±adirlo a nuestra lista, ademas de ir acumulando los promedios
"""

for x,y,z in zip(nombres,notas1,notas2):
    aux = {"Nombre":x , "Nota Total": int(y) + int(z)}
    estructura.append(aux)
    promedio += aux["Nota Total"]
    
"""Dividimos el promedio con la cantidad de elementos, o sea la cantidad total de estudiantes"""

promedio = promedio / len(estructura)

"""Hacemos un print y recorremos toda la estructura para ver que estudiante esta por debajo del promedio, y si es el caso
se imprime en pantalla"""

print(estructura)

for x in estructura:
    if(x["Nota Total"]<promedio):
        print(f'El estudiante {x["Nombre"]}esta por debajo del promedio.')
```
